### Saving a Colab Notebook to GitHub

1.  **Open your notebook** in Google Colab.
2.  Go to **File > Save a copy to GitHub...**.
3.  A new window or tab will open, prompting you to **authorize Colab to access your GitHub account** if you haven't already.
4.  Once authorized, you can select the **GitHub repository** and **branch** where you want to save your notebook. You can also provide a **commit message**.
5.  Click **OK** to save the notebook. This will create a new commit in your chosen repository with the `.ipynb` file.

### Cloning a GitHub Repository into Colab

To work with an existing GitHub repository in Colab, you can clone it directly into your Colab environment. This allows you to access and modify files within that repository.

1.  **Get the clone URL** of your GitHub repository (e.g., `https://github.com/your-username/your-repo.git`).
2.  In a Colab code cell, use the `git clone` command. You might need to provide credentials if it's a private repository.
